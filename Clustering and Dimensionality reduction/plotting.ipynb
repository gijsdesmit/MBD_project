```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('dbs_results.csv') # Contains 

#print(df.head())

valid_clusters = df[df['prediction'] != -1]
print(f'Number of valid clusters: {len(valid_clusters["prediction"].unique())}')
#percentage of outliers
print(f'Percentage of outliers: {len(df[df["prediction"] == -1]) / len(df) * 100:.2f}%')
score = silhouette_score(valid_clusters[['Dimension 1', 'Dimension 2']], valid_clusters['prediction'])
print(f'Silhouette score: {score:.3f}')
```

```python
plt.figure(figsize=(10, 10))

plt.scatter(df['Dimension 1'], df['Dimension 2'], s=3)

#plt.legend()
plt.xlabel('t-SNE 1', fontsize=14)
plt.ylabel('t-SNE 2', fontsize=14)
plt.title('t-SNE applied on the semantic vector space', fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=14)
#save plot
plt.savefig('images/tsne.png', dpi=450, bbox_inches='tight')
plt.show()


```

```python
plt.figure(figsize=(10, 10))
for cluster in df['prediction'].unique():
    d = df[df['prediction'] == cluster]
    plt.scatter(d['Dimension 1'], d['Dimension 2'], label=f'Cluster {cluster}', s=3)

#plt.legend()
plt.tick_params(axis='both', which='major', labelsize=14)
plt.xlabel('t-SNE 1', fontsize=14)
plt.ylabel('t-SNE 2', fontsize=14)
plt.title('DBSCAN Clustering applied to reduced dimensionality space', fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
for cluster in df['prediction'].unique():
    d = df[df['prediction'] == cluster]
    if cluster == -1:
        plt.scatter(d['Dimension 1'], d['Dimension 2'], color='black', marker='x', label='Outliers', s=4.7, linewidth=0.7)
    else:
        plt.scatter(d['Dimension 1'], d['Dimension 2'], s=3)
        #plt.scatter(d['Dimension 1'], d['Dimension 2'], label=f'Cluster {cluster}', s=3)

plt.tick_params(axis='both', which='major', labelsize=14)
plt.xlabel('t-SNE 1', fontsize=14)
plt.ylabel('t-SNE 2', fontsize=14)
plt.title('DBSCAN Clustering on t-SNE space', fontsize=18)
plt.legend(loc='upper right', fontsize=14)
plt.savefig('images/dbscan.png', dpi=450, bbox_inches='tight')
plt.show()
```

```python
# c = 0, "blindness"
# c = 1, "Login"
# c = 2, "Brewing (Beer)"
# c = 3, "Technology"
# c = 4, "Military history"
# c = 5, "ATPM"
# c = 6, "Christianity"
# c = 7, "Travel"
# c = 8, "Art and Education"
# c = 9, "Cinema"
# c = 11 "Software Development"
# c = 12 "Music"

cluster_dict = {0: "ACB", 1: "Login", 2: "Brewing (Beer)", 3: "Technology", 4: "Military history", 5: "ATPM", 6: "Christianity", 7: "Travel", 8: "Art and Education", 9: "Cinema", 11: "Software Development", 12: "Music"}
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# Normalize cluster numbers to [0,1]
norm = plt.Normalize(min(cluster_dict.keys()), max(cluster_dict.keys()))

# Create a color map with a unique color for each cluster in the dictionary
color_map = {cluster: cm.rainbow(norm(cluster)) for cluster in cluster_dict.keys()}

# Plot all the data in a default color (for example, gray)
plt.figure(figsize=(10, 10))
plt.scatter(df['Dimension 1'], df['Dimension 2'], color='gray', s=3)

# Plot each cluster in the dictionary with its unique color
for cluster, color in color_map.items():
    if cluster == -1:
        continue
    
    d = df[df['prediction'] == cluster]
    plt.scatter(d['Dimension 1'], d['Dimension 2'], color=color, label=cluster_dict[cluster], s=3)

plt.tick_params(axis='both', which='major', labelsize=14)
plt.xlabel('t-SNE 1', fontsize=14)
plt.ylabel('t-SNE 2', fontsize=14)
plt.title('DBSCAN Clustering highlighted clusters (outliers removed)', fontsize=18)
legend=plt.legend(loc='upper right', fontsize=12)#  bbox_to_anchor=(1.34, 1),
for handle in legend.legend_handles:
    handle._sizes = [35]  # adjust as necessary
    
plt.savefig('images/dbscan_highlighted.png', dpi=450, bbox_inches='tight')
plt.show()
```
